## Hipotesa
Setelah ditemukan data diatas saya ingin membuat hipotesis berupa:
- H0: Î¼ YGF == Î¼ YGM Maksudnya adalah jika nilai Rata rata sample Female == Rata-rata sample Male

- H1: Î¼ YGF != Î¼ YGM Maksudnya adalah jika nilai Rata rata sample Female != Rata-rata sample Male


```python
t_stat, p_val = stats.ttest_ind(ygf['gross income'], ygm ['gross income'])
# T-test itu secara default menggunakan 2 tail, jadi tidak perlu dibagi 2/0,5
print('P-value:',p_val)       #Menampilkan P-Value
print('t-statistics:',t_stat)   #Menampilkan T-Statistic
```

Dari sini bisa disimpulkan bahwa H0 diterima , akan tetapi saya akan mencoba membuktikannya lagi berdasarkan grafik, berikut simulasinya:

```python
# simulation
ygf_pop = np.random.normal(ygf['gross income'].mean(),ygf['gross income'].std(),10000)
ygm_pop = np.random.normal(ygm['gross income'].mean(),ygm['gross income'].std(),10000)

# confidence interval based on gen 4
# menggunakan critical value sebesar 0.05 ,ketika 2 tailed confidence interval menjadi 0.95
ci = stats.norm.interval(0.95, ygf['gross income'].mean(), ygf['gross income'].std())

plt.figure(figsize=(16,5))
sns.distplot(ygf_pop, label='Yangoon Female Gross Income ',color='blue')
sns.distplot(ygm_pop, label='Yangoon Male Gross Income ',color='red')

# mean
plt.axvline(ygf['gross income'].mean(), color='blue', linewidth=2, label='Yangoon Female Gross Income mean')
plt.axvline(ygm['gross income'].mean(), color='red',  linewidth=2, label='Yangoon Male Gross Income mean')

plt.axvline(ci[1], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%')
plt.axvline(ci[0], color='green', linestyle='dashed', linewidth=2, label='confidence threshold of 95%')

# alternative hypotesis
plt.axvline(ygm_pop.mean()+t_stat*ygm_pop.std(), color='black', linestyle='dashed', linewidth=2, label = 'Alternative Hypothesis')
plt.axvline(ygm_pop.mean()-t_stat*ygm_pop.std(), color='black', linestyle='dashed', linewidth=2)

plt.legend()
print(ci[0])
```

### Kesimpulan

Berdasarkan hasil analisis di atas bisa disimpulkan bahwa:
- Nilai P-value > Critical Value yaitu 0.16261180081238572 >  5% 
- Berdasarkan dari data diatas bisa dilihat bahwa nilai dari Alternatif Hypothesis (X) didalam batas Critical Value (cv) sehingga dapat disimpulkan bahwa nilai Rata rata sample Female memiliki buktinya yang cukup sama dengan Rata-rata sample Male 
- Jadi hipotesis yang dapat disimpulkan yaitu : 
- H0: Î¼ YGF == Î¼ YGM Maksudnya adalah jika nilai Rata rata sample Female == Rata-rata sample Male

Itu menandakan bahwa Rata-rata gross income antara Gender Laki-laki dengan Gender Female memiliki kesamaan, tetapi kalau dilihat dari nilai Mean yang dimiliki keduanya, tetap Gender Female yang lebih besar dibanding Gender Male (F> M).
